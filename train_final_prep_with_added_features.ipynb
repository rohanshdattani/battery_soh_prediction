```python
cap_fade = []
for j in range(df2.shape[0]):
    if j == 0:
        cap_fade.append([0])
    elif df2['cell_id'].iloc[j] == df2['cell_id'].iloc[j-1] and j <=(df2.shape[0]-2):
        cap_fade.append(df2.iloc[j+1,-1] - df2.iloc[j,-1])
    else:
        cap_fade.append(df2.iloc[j,-1] - df2.iloc[j-1,-1])

cap_fade
    
        
```

```python
df = df.sort_values(['cell_id','cycle'])

df['capacity_fade'] = df.groupby('cell_id')['capacity'].diff()

df['capacity_fade'] = df['capacity_fade'].fillna(0)

df[['cell_id','cycle','capacity','capacity_fade']].head(20)
```

```python
df = df.drop(columns = ['charge_max', 'charge_min', 'time_max', 'time_min'])
```

```python
df
```

```python

df['cycle_number'] = df['cycle'].str.extract(r'(\d+)').astype(int) // 100


df = df.sort_values(['cell_id','cycle_number']).reset_index(drop=True)
```

```python
df
```

```python

initial_capacity = df.groupby('cell_id')['capacity'].transform('first')


df['SOH'] = df['capacity'] / initial_capacity


df[['cell_id','cycle','capacity','SOH']].head(20)
```

```python
df = df.drop(columns = ['cycle'])
```
