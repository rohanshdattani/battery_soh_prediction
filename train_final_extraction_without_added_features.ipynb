```python
# %%
import scipy.io as sio
import pandas as pd
import numpy as np
import pandas as pd

# Set options to display all rows and columns


data = sio.loadmat('Oxford_Battery_Degradation_Dataset_1.mat')
data.keys()

data['Cell1'][0][0].dtype.names


rows = []
q = []
v = []
t = []
T = []
cell_id = []
phase_name = []
cycle_name = []

for i in range(1,7):
        cell= data['Cell'+str(i)][0][0]
    
    

    
        cycle_names = cell.dtype.names
        
        for cycle in cycle_names:
            cycle_data = cell[cycle][0][0]
            phase_names = cycle_data.dtype.names
            for phase in phase_names:
                phase_data = cycle_data[phase][0][0]
                for j in range(len(phase_data['q'])):
                    q.append(phase_data['q'][j][0])
                    v.append(phase_data['v'][j][0])
                    t.append(phase_data['t'][j][0])
                    T.append(phase_data['T'][j][0])
                    cell_id.append(i)
                    phase_name.append(phase)
                    cycle_name.append(cycle)







    

#print(q)
#print(len(q))
#print(len(v))
#print(len(t))
#print(len(T))
#print(len(cell_id))
#print(len(cycle_name))
#print(len(phase_name))

# %%
df = pd.DataFrame({
    'cell_id': cell_id,
    'cycle': cycle_name,
    'phase': phase_name,
    'charge': q,
    'time': t,
    'temperature': T,
    'voltage': v

}) 
# %%
df.shape
# %%
```
