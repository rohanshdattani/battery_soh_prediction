```python

import scipy.io as sio
import pandas as pd
import numpy as np
import pandas as pd



data = sio.loadmat('Oxford_Battery_Degradation_Dataset_1.mat')
data.keys()

data['Cell1'][0][0].dtype.names


rows = []
q = []
v = []
t = []
T = []
cell_id = []
phase_name = []
cycle_name = []

for i in range(7,9):
        cell= data['Cell'+str(i)][0][0]
    
    

    
        cycle_names = cell.dtype.names
        
        for cycle in cycle_names:
            cycle_data = cell[cycle][0][0]
            phase_names = cycle_data.dtype.names
            for phase in phase_names:
                phase_data = cycle_data[phase][0][0]
                for j in range(len(phase_data['q'])):
                    q.append(phase_data['q'][j][0])
                    v.append(phase_data['v'][j][0])
                    t.append(phase_data['t'][j][0])
                    T.append(phase_data['T'][j][0])
                    cell_id.append(i)
                    phase_name.append(phase)
                    cycle_name.append(cycle)







    




df = pd.DataFrame({
    'cell_id': cell_id,
    'cycle': cycle_name,
    'phase': phase_name,
    'charge': q,
    'time': t,
    'temperature': T,
    'voltage': v

}) 



features = df.groupby(["cycle", "cell_id"]).agg({
    "voltage": ["mean","std","max","min"],
    "temperature": ["mean","max"],
    "charge": ["max","min"],
    "time": ["max","min"]
})
features.columns = ['_'.join(col) for col in features.columns]
features
```

```python
features['capacity'] = features['charge_max'] - features['charge_min']
```

```python
features.drop(columns = ["charge_min"])
```

```python
features['v_range'] = features['voltage_max'] - features['voltage_min']
```

```python
features = features.drop('v_range', axis = 1)
```

```python
features = features.sort_values(by=["cell_id","cycle"])
```

```python

features = features.sort_values(['cell_id','cycle'])


features['capacity_fade'] = features.groupby('cell_id')['capacity'].diff()


features['capacity_fade'] = features['capacity_fade'].fillna(0)

features[['cell_id','cycle','capacity','capacity_fade']].head(20)
```

```python
features = features.drop(columns = ['charge_max', 'charge_min', 'time_max', 'time_min'])
```

```python

features['cycle_number'] = features['cycle'].str.extract(r'(\d+)').astype(int) // 100

features = features.sort_values(['cell_id','cycle_number']).reset_index(drop=True)
```
